```python
#Pandas is a library having the functions related to the database
import pandas as pd
```

```python
#Assigning a variable "dataset" agaisnt the CSV file need to be taken
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#Recall/print the variable details
dataset
```

```python
#extract the needed column, in "dataset" variable contain whole data, raising a new variable as "independent" and calling 
#the relavant column "YearsExperience" which is input data
independent=dataset[["YearsExperience"]]
independent
```

```python
#print or recall the variable
independent
```

```python
#extract the needed column, in "dataset" variable contain whole data, raising a new variable as "dependent" and calling 
#the relavant column "Salary" which is output data
dependent=dataset[["Salary"]]
dependent
```

```python
#sklearn= split the the train and test sets, [sklearn.model dt operator reffering here like class and functions this lead 
#save the memory space rather than recall all library]
from sklearn.model_selection import train_test_split

#Recall the function and update the parameter in the brackets, train and tests for indpendent & train & test for dependent 
# to split and visible
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python

#AI Model Creation Part (Learning Phase) =>
#Call the Linear Regression Function using "sklearn.linear_model
from sklearn.linear_model import LinearRegression

#Creating variable as "regressor", and assiginig "LineaRegression()" as like class which already inclueds functions
regressor=LinearRegression()

#Calling the function "fit" which inside of regessor, and adding the inputs and outputs for the training sets for substitute
#like apply the data for the formuals (Y=MX+C)
regressor.fit(X_train,y_train)
```

```python
#weight is calculated from .coef_ formula thus using here to call
weight=regressor.coef_
weight
```

```python
#checking the bais variable by recalling ".intercept_" function
bais=regressor.intercept_
bais
```

```python
#Evaluvate the model
#Creating a variable "y_pred" call the predict function includes in the regressor 
# like calling from regressor calss, function predict
#predict the X_test data, which need to be evaluvated cross the check test data which isn't included in training set
y_pred=regressor.predict(X_test)
```

```python
#Calculate the r2 value
from sklearn.metrics import r2_score
#creating a variable "r_score" and substitute to the formula
#apply the actual Y data , and Y predicted Data
r_score=r2_score(y_test,y_pred)
```

```python
#print the r_score value, value is merely one which means model is performing well enough
r_score
```

```python
#saving the model which can be done by "pickle" 
import pickle

#creating variable as "filename" and give the name
#pickle extension is ".sav" so saving accordingly is creating onle file name
filename="finalized_model_linear.sav"
```

```python
#save the model in to the file 
# regressor where all model is saved, wb means write binary

pickle.dump(regressor,open(filename,'wb'))
```

```python
# load the saved Model
#Creating a variable "loaded_model" , call the function "pickle.load" then "open" and path of the file
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))

#creating a variable "results" to call the "loaded_model"
result=loaded_model.predict([[13]])
```
